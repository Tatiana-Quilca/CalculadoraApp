```python
# -*- coding: utf-8 -*-
"""
Created on Tue Jun 17 08:55:07 2025

@author: IVAN
"""

import tensorflow as tf
# Importaciones para DCGAN: Conv2D y Conv2DTranspose
from tensorflow.keras.layers import Dense, LeakyReLU, BatchNormalization, Reshape, Flatten, Conv2D, Conv2DTranspose, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt

# Configuración inicial
np.random.seed(42)
tf.random.set_seed(42)

# Hiperparámetros
latent_dim = 100 # Reducido un poco para DCGANs, 100-128 es común
img_shape = (28, 28, 1) # Forma de las imágenes MNIST
epochs = 5000 # Aumentado significativamente para ver buenos resultados con DCGAN
batch_size = 64
lr_D = 0.0002 # Learning rate para el Discriminador (común en DCGANs)
lr_G = 0.0002 # Learning rate para el Generador (común en DCGANs)
beta_1 = 0.5 # Parámetro para Adam

# Cargar y preprocesar datos MNIST
(X_train, y_train), (_, _) = mnist.load_data()
# Normalizar imágenes a [-1, 1]
X_train = (X_train.astype(np.float32) - 127.5) / 127.5
X_train = np.expand_dims(X_train, axis=3) # Añadir dimensión de canal (28, 28, 1)


# Seleccionar algunas imágenes aleatorias (ej. 9, 16) de ejemplo
num_images = 16
random_indices = np.random.randint(0, X_train.shape[0], num_images)
sample_images_mnist = X_train[random_indices]
sample_labels = y_train[random_indices]

# Configurar el grid de visualización
rows = 4
cols = 4
fig, axes = plt.subplots(rows, cols, figsize=(8, 8))

# Mostrar las imágenes (re-escalar para mostrar si están en [-1, 1])
for i, ax in enumerate(axes.flat):
    img_to_show = (sample_images_mnist[i] + 1) / 2 # Re-escalar para visualización
    ax.imshow(img_to_show.reshape(28, 28), cmap='gray') # 'gray' para escala de grises
    ax.set_title(f"Label: {sample_labels[i]}", fontsize=10)
    ax.axis('off') # Ocultar ejes

plt.suptitle('Ejemplos del Dataset MNIST', fontsize=14)
plt.tight_layout()
plt.show()


# **Generador (DCGAN)**

def build_generator():
    model = Sequential([
        # Primer Dense para expandir el ruido latente a un volumen
        Dense(7 * 7 * 256, input_dim=latent_dim),
        LeakyReLU(alpha=0.2),
        BatchNormalization(momentum=0.8),
        Reshape((7, 7, 256)), # Reshape a un volumen 3D para Conv2DTranspose

        # Capa Conv2DTranspose 1: Sube la resolución (7x7 -> 14x14)
        Conv2DTranspose(128, kernel_size=4, strides=2, padding='same', use_bias=False), # use_bias=False con BN
        LeakyReLU(alpha=0.2),
        BatchNormalization(momentum=0.8),

        # Capa Conv2DTranspose 2: Sube la resolución (14x14 -> 28x28)
        Conv2DTranspose(64, kernel_size=4, strides=2, padding='same', use_bias=False),
        LeakyReLU(alpha=0.2),
        BatchNormalization(momentum=0.8),

        # Capa de salida: Genera la imagen final de 28x28x1
        # La última capa convolucional no usa BatchNormalization
        Conv2D(1, kernel_size=3, padding='same', activation='tanh') # Tanh para salida [-1, 1]
    ])
    return model


# **Discriminador (DCGAN)**

def build_discriminator():
    model = Sequential([
        # Capa Conv2D 1: Recibe 28x28x1, reduce el tamaño y aumenta los filtros
        Conv2D(64, kernel_size=3, strides=2, input_shape=img_shape, padding='same'),
        LeakyReLU(alpha=0.2),
        Dropout(0.25), # Ayuda a prevenir el sobreajuste

        # Capa Conv2D 2: Reduce más el tamaño
        Conv2D(128, kernel_size=3, strides=2, padding='same', use_bias=False),
        LeakyReLU(alpha=0.2),
        BatchNormalization(momentum=0.8),
        Dropout(0.25),

        # Capa Conv2D 3: Reduce aún más
        Conv2D(256, kernel_size=3, strides=2, padding='same', use_bias=False),
        LeakyReLU(alpha=0.2),
        BatchNormalization(momentum=0.8),
        Dropout(0.25),

        Flatten(), # Aplanar la salida convolucional para la capa Dense
        Dense(1, activation='sigmoid') # Salida binaria (real/falso)
    ])
    return model

# Compilar el discriminador
discriminator = build_discriminator()
discriminator.compile(
    optimizer=Adam(learning_rate=lr_D, beta_1=beta_1), # Usar lr_D
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Compilar el generador
generator = build_generator()

# Congelar el discriminador durante el entrenamiento del generador
discriminator.trainable = False

# Modelo GAN (Generador -> Discriminador)
gan = Sequential([generator, discriminator])
gan.compile(
    optimizer=Adam(learning_rate=lr_G, beta_1=beta_1), # Usar lr_G
    loss='binary_crossentropy'
)

# **Entrenamiento (con visualización)**

def train_gan(epochs, batch_size, sample_interval=50):
    # Etiquetas para datos reales y falsos (etiquetas suavizadas para reales)
    real = np.ones((batch_size, 1)) * 0.9 # Usar 0.9 para etiquetas reales
    fake = np.zeros((batch_size, 1))

    for epoch in range(epochs):
        # ---------------------
        # Entrenar Discriminador
        # ---------------------
        idx = np.random.randint(0, X_train.shape[0], batch_size)
        real_imgs = X_train[idx]

        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        gen_imgs = generator.predict(noise, verbose=0)

        # Entrenar el discriminador con imágenes reales
        d_loss_real = discriminator.train_on_batch(real_imgs, real)
        # Entrenar el discriminador con imágenes falsas
        d_loss_fake = discriminator.train_on_batch(gen_imgs, fake)
        # Pérdida promedio del discriminador
        d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

        # -----------------
        # Entrenar Generador
        # -----------------
        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        # El generador intenta engañar al discriminador haciendo que las imágenes falsas se clasifiquen como reales
        g_loss = gan.train_on_batch(noise, real)

        # Progreso
        if epoch % sample_interval == 0:
            print(f"Época {epoch} [D loss: {d_loss[0]:.4f}, acc.: {100*d_loss[1]:.2f}%] [G loss: {g_loss:.4f}]")
            sample_images(epoch)

# Función para mostrar las imágenes generadas
def sample_images(epoch):
    r, c = 4, 4 # Grid de 4x4 imágenes
    noise = np.random.normal(0, 1, (r * c, latent_dim))
    gen_imgs = generator.predict(noise, verbose=0)

    # Re-escalar imágenes de [-1, 1] a [0, 1] para visualización
    gen_imgs = 0.5 * gen_imgs + 0.5

    fig, axs = plt.subplots(r, c)
    cnt = 0
    for i in range(r):
        for j in range(c):
            # Asegúrate de que la imagen sea 2D para imshow si es de 1 canal
            axs[i,j].imshow(gen_imgs[cnt].reshape(img_shape[0], img_shape[1]), cmap='gray')
            axs[i,j].axis('off')
            cnt += 1

    plt.suptitle(f'Época {epoch}', fontsize=12)
    plt.tight_layout()
    plt.show() # Muestra la figura en pantalla
    plt.close() # Cierra la figura para liberar memoria

# Iniciar entrenamiento
train_gan(epochs=epochs, batch_size=batch_size, sample_interval=200) # Aumentar sample_interval si epochs es muy grande
```
